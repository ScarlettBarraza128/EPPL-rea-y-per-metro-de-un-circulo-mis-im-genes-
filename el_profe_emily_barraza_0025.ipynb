```python
# Calcular area y Perrmetro de un circulo
# variables A, P, r
# formulas A = pi * r^2 y P = 2 * pi * r
# capturar el valor de r
import math
print("Calcular el area y el perimetro de un circulo")
r = float(input("Ingrese el valor del radio: "))
A = math.pi*r**2
P = 2*math.pi*r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Emily Barraza NL = 7")
```
